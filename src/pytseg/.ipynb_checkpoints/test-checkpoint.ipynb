```python
import numpy as np
from plot import plot
from data import test_fun

x, t = test_fun()
```

```python
from plot import plot

plot(x, t=t)

from seg import normalize

x,t = normalize(x, t)
plot(x, t=t)
```

```python
from seg import cut

s = cut(x, .95, 3)
plot(x, s=s, figure_kwarg_dict=dict(figsize=(10,5)))
```

```python
from seg import segment_stationarity, segment_clustering

l = segment_stationarity(x, s, threshold=.001)
plot(x, s=s, l=l, figure_kwarg_dict=dict(figsize=(10,5)))
        
l = segment_clustering(x, s, n_clusters=3, random_state=0, feature_maps=[np.mean, np.std])
plot(x, s=s, l=l, figure_kwarg_dict=dict(figsize=(10,5)))
```

```python
l = segment_clustering(x, s, n_clusters=5, random_state=0, feature_maps=[np.std])
plot(x, s=s, l=l, figure_kwarg_dict=dict(figsize=(20,5)))
```

```python
l = segment_clustering(x, s, n_clusters=5, random_state=0, feature_maps=[np.mean])
plot(x, s=s, l=l, figure_kwarg_dict=dict(figsize=(10,5)))
```

```python
def fm(x):
    return np.max(x) - np.min(x)

l = segment_clustering(x, s, n_clusters=3, random_state=0, feature_maps=[fm])
plot(x, s=s, l=l, figure_kwarg_dict=dict(figsize=(10,5)))
```

```python
def fm1(x):
    return np.max(x)
def fm2(x):
    return np.min(x)

l = segment_clustering(x, s, n_clusters=3, random_state=0, feature_maps=[fm1,fm2])
plot(x, s=s, l=l, figure_kwarg_dict=dict(figsize=(10,5)))
```

```python
import numpy as np
from data import test_fun

t = np.arange(7000)

x1, _ = test_fun(t)
x2, _ = test_fun(t, 6=1/50, 9=2000)
x3, _ = test_fun(t, 1=2, 2=1/100, 9=3500)

X = np.vstack([x1,x2,x3]).T

X.shape, t.shape
```

```python
from plot import plot_multi

plot_multi(X, t=t, figure_kwarg_dict=dict(figsize=(10,5)), label='x')

from seg_multi import normalize_multi

X,t = normalize_multi(X, None)
plot_multi(X, t=t, figure_kwarg_dict=dict(figsize=(10,5)), label='x')
```

```python
from seg_multi import cut_multi

S = cut_multi(X, .95, 3)
plot_multi(X, S=S, figure_kwarg_dict=dict(figsize=(10,5)))
```

```python
from seg_multi import segment_stationarity_multi

s, l = segment_stationarity_multi(X, S, threshold=.001)

plot_multi(X, t=t, S=[s]*X.shape[1], L=[l]*X.shape[1], figure_kwarg_dict=dict(figsize=(10,5)), plot_kwarg_dict=dict(lw=2))
```
